### VAR estimation of correlated exogenous states

Exogenous states Markov process:

* $VAR(1)$ process:

    \begin{equation*}
        \begin{bmatrix}
        \ln{(y_t^{T})} \\[0.3cm]
        \ln{(s_t)} 
        \end{bmatrix}
        = \mathbf{A}
        \begin{bmatrix}
        \ln{(y_{t-1}^{T})} \\[0.3cm]
        \ln{(s_{t-1})}
        \end{bmatrix} + \boldsymbol{\varepsilon}_t;\;\
        ;\;\;\hat{\boldsymbol{\Sigma}} = \frac{\sum_{t=2}^{T}\,\hat{\boldsymbol{\varepsilon}}_t\hat{\boldsymbol{\varepsilon}}_t'}{T} 
    \end{equation*}

```julia
using QuantEcon, LinearAlgebra, Statistics, Queryverse
using DataFrames, VectorAutoregressions, JLD
```

```julia
# Load data and set VAR(1) Parameters
dataSpain = load("./data/Var_GDPT_R_Spain.csv") |>
    @select(:l_gdp_T_cyc_hp,:l_dev_R_SP) |>
    DataFrame

dataItaly = load("./data/Var_GDPT_R_Italy.csv") |>
    @select(:l_gdp_T_cyc_hp,:l_dev_R_IT) |>
    DataFrame

dataPortugal = load("./data/Var_GDPT_R_Portugal.csv") |>
    @select(:l_gdp_T_cyc_hp,:l_dev_R_PO) |>
    DataFrame

y_Spain = [dataSpain.l_gdp_T_cyc_hp dataSpain.l_dev_R_SP]
y_Italy = [dataItaly.l_gdp_T_cyc_hp dataItaly.l_dev_R_IT]
y_Portugal = [dataPortugal.l_gdp_T_cyc_hp dataPortugal.l_dev_R_PO]

intercept = false
p = 1
```

```julia
# Spain: Fit VAR(1) to data
VAR_Spain = VAR(y_Spain, p, intercept)
VAR_Spain.β
```

```julia
JLD.save("./data/A_Spain.jld", "A_Spain", VAR_Spain.β)
VAR_Spain.Σ
```

```julia
JLD.save("./data/Sigma_Spain.jld", "Sigma_Spain", VAR_Spain.Σ)
```

```julia
# Italy: Fit VAR(1) to data
VAR_Italy = VAR(y_Italy, p, intercept)
VAR_Italy.β
```
