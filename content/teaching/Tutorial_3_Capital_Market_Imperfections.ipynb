1. **Financial Autarky in the intertemporal model of the SOE**. Consider the 2-period economy under Financial Autarky covered in class. Assume that preferences of the consumers are of the form

    $$ U_{1} =  \log{(c_{1})}+\beta\,\log{(c_{2})}$$

    - Starting from the stationary equilibrium such that $y_{1}=y_{2}=y^{H}$, compute the equilibrium real domestic interest rate ($r^{d}$).

        Remember that under the assumptions: $i)$ $b_{0}=b_{0}^{d}=0$, and $ii)$ the domestic bond is in *zero-net-supply*, the equilibrium of the domestic bond market implies:

        $$
            b_{1}^{d} = 0.
        $$

        From the flow budget constraints of the consumers, *financial autarky* immediately implies:

        \begin{gather*}
        c_{1} = y_{1}=y^{H}\\
        c_{2} = y_{2}=y^{H}.
        \end{gather*}
        
        From the Euler equation we can determine the domestic interest rate $r^{d}$ that clears the domestic bond market, i.e.:

        \begin{gather*}
        u^{\prime}(\underbrace{y^{H}}_{=c_{1}}) = \beta(1+r^{d})u^{\prime}(\underbrace{y^{H}}_{=c_{2}})\\
        1+r^{d} = \frac{1}{\beta}\frac{u^{\prime}(y^{H})}{u^{\prime}(y^{H})} \color{red}=\frac{1}{\beta}
        \end{gather*}
        

1. ...

    - Now assume that there is a transitory shock such that, unexpectedly, the output endowment of period 1 falls to $y_{1}=y^{L}<y^{H}$. Compute the equilibrium real domestic interest rate $(r^{d})$ of this economy and determine the new perfect foresight equilibrium paths of consumption and the current account balance and the equilibrium value of welfare $U_{1}$. Is the economy better-off or worse-off compared to a situation in which the economy would have access to the international capital market? (*Hint: to make the comparison, you need to remember (or compute) the perfect foresight consumption equilibrium path when the economy has perfect access to the international capital market under the assumption that $\beta=\frac{1}{1+r}$*)

        Once the transitory income shock is realised, financial autarky and the equilibrium in the domestic bond market imply that the new perfect foresight equilibrium path is:

        \begin{gather*}
        c_{1} = y_{1}=y^{L}\\
        c_{2} = y_{2}=y^{H}.
        \end{gather*}
        
        Once again from the Euler equation it is possible to determine the value of the domestic interest rate that clears the domestic bond market in the face of the transitory income shock, i.e.:

        \begin{gather*}
        u^{\prime}(\underbrace{y^{L}}_{=c_{1}}) = \beta(1+r^{d})u^{\prime}(\underbrace{y^{H}}_{=c_{2}})\\
        1+r^{d} = \frac{1}{\beta}\frac{u^{\prime}(y^{L})}{u^{\prime}(y^{H})}
        \end{gather*}
        
        
        Now, from the explicit for of $U_{1}$ we know that $u(c)=\log{(c)}$. Therefore, $u^{\prime}(c)=\frac{1}{c}$ which implies that the gross equilibrium domestic real interest rate is now:

        $$
            \color{red}1+r^{d} = \frac{1}{\beta}\overbrace{\left(\frac{y^{H}}{y^{L}}\right)}^{>1} > \frac{1}{\beta}
        $$

        Since before and after the transitory shock the SOE will be characterised by an *Autarky* equilibrium, the perfect foresight equilibrium paths of the current account balance will always be:

        $$
            ca_{1}=ca_{2}=0.
        $$

        Welfare after the transitory shock will be fully determined by the equilibrium path of consumption:

        $$
            \color{red}U_{1} = \log{(y^{L})}+\beta\,\log{(y^{H})}.
        $$

        We can compare this level of welfare to the one which consumers will get if they were able to participate in the international financial market to smooth consumption. Remember that under the assumption that $\beta=\frac{1}{1+r}$, after the transitory shock the constant path of consumption in this case would be 
        
        $$
            c=\left(\frac{1+r}{2+r}\right)\left(y^{L}+\frac{y^{H}}{1+r}\right)\stackrel{\beta=\frac{1}{1+r}}{\Longrightarrow}\color{blue}c=\frac{1}{1+\beta}y^{L}+\frac{\beta}{1+\beta}y^{H}, 
        $$
        
        that would imply a level of welfare:

        $$
            \log{\left(\frac{1}{1+\beta}y^{L}+\frac{\beta}{1+\beta}y^{H}\right)}+\beta\,\log{\left(\frac{1}{1+\beta}y^{L}+\frac{\beta}{1+\beta}y^{H}\right)}=\color{blue}(1+\beta)\log{\left(\frac{1}{1+\beta}y^{L}+\frac{\beta}{1+\beta}y^{H}\right)}.
        $$

        Now, under the conjecture that welfare is higher when the SOE has access to the international capital markets, the comparison would depend on the inequality:

        $$
            (1+\beta)\log{\left(\frac{1}{1+\beta}y^{L}+\frac{\beta}{1+\beta}y^{H}\right)} > \log{(y^{L})}+\beta\,\log{(y^{H})}
        $$

        that once we divide both sides by $1+\beta$ can be rewritten as:

        $$
            \color{red}\log{\left(\frac{1}{1+\beta}y^{L}+\frac{\beta}{1+\beta}y^{H}\right)} > \frac{1}{1+\beta}\log{(y^{L})} +\frac{\beta}{1+\beta}\log{(y^{H})}
        $$

        which holds true given the strict concavity of the $\log{(\cdot)}$ function (see graph on the lecture slides!), meaning that the SOE is worse-off compared to a situation in which it would have access to the international financial market.

2. **Example of an incomplete-markets economy (intertemporal model of the SOE)**. Consider the 2-period economy with uncertainty covered in class. Assume that preferences of the consumers are such that

    $$ u(c) = \frac{c^{1-\theta}-1}{1-\theta} $$
    
    - Show that the coefficient of relative risk aversion $R(c)=-c\frac{u^{\prime\prime}(c)}{u^{\prime}(c)}$ is equal to $\theta$.

        To compute $R(c)$ we need to compute the first and second derivatives of the $u(c)$ function, i.e.:

        \begin{gather}
        u^{\prime}(c) = c^{-\theta}\\
        u^{\prime\prime}(c) = -\theta c^{-(\theta+1)}
        \end{gather}

        Therefore,

        $$
        R(c) = -c\frac{-\theta c^{-(\theta+1)}}{c^{-\theta}}=\theta\frac{c^{-\theta}}{c^{-\theta}}\color{red}=\theta
        $$

2. ...    
    - Are these consumers *prudent*?

        Now we need to compute $u^{\prime\prime\prime}(c)$:

        $$
        u^{\prime\prime\prime}(c) = \theta(\theta+1)c^{-(\theta+2)}=\frac{\theta(\theta+1)}{c^{(\theta+2)}}\color{red}>0.
        $$

        Since $u^{\prime\prime\prime}(c)>0$ these consumers are *prudent* consumers.

2. ...

    - Plot the equilibrium values $c_{1},c_{2}^{H},c_{2}^{L},ca_{1}$ as a function of $\theta$.

        Under the assumption that $\beta=\frac{1}{1+r}$, let us start from the general equilibrium Euler equation of the 2-period SOE model under uncertainty covered in class:

        $$
            u^{\prime}(c_{1}) = pu^{\prime}(c_{2}^{H})+(1-p)u^{\prime}(c_{2}^{L})
        $$

        By using the first derivative computed above, the explicit form of the Euler equation is:

        $$
            \color{red}c_{1}^{-\theta} = p(c_{2}^{H})^{-\theta}+(1-p)(c_{2}^{L})^{-\theta}
        $$

        and the general equilibrium allocation depends also on the two possible (state-of-the-world) IBCs, from which:

        \begin{align*}
        c_{2}^{H} &= (1+r)(y_{1}-c_{1}) + y_{2}^{H}\\
        c_{2}^{L} &= (1+r)(y_{1}-c_{1}) + y_{2}^{L}, 
        \end{align*}

        and the equations for the current account balances:

        \begin{align*}
        ca_{1} &= y_{1} - c_{1}\\
        ca_{2}^{H} &= y_{2}^{H} - c_{2}^{H}\\
        ca_{2}^{L} &= y_{2}^{L} - c_{2}^{L}
        \end{align*}
        
        An attempt to solve using pen and paper for the equilibrium allocation could start from substituting the IBCs into the Euler equation to solve for $c_{1}$, i.e.:

        $$
        \color{blue}c_{1}^{-\theta} = p(\overbrace{(1+r)(y_{1}-c_{1}) + y_{2}^{H}}^{=c_{2}^{H}})^{-\theta}+(1-p)(\underbrace{(1+r)(y_{1}-c_{1}) + y_{2}^{L}}_{=c_{2}^{L}})^{-\theta}. 
        $$

        Unfortunately, the non-linearities characterising this equation make it impossible to solve explicitly for $c_{1}$. To get a solution we would need to take the equation to the computer, and solve it using a numerical method to find the zero-root of the equation:

        $$
         0 \approx c_{1}^{-\theta} - p((1+r)(y_{1}-c_{1}) - y_{2}^{H})^{-\theta} - (1-p)((1+r)(y_{1}-c_{1}) + y_{2}^{L})^{-\theta}
        $$

        Below, you can find a piece of code in python to solve for the general equilibrium allocation. It depends on two *functions* which define first, a model for the SOE in which consumers have certainty equivalence-behaviour (```certainty_eq_SOE()```), and second the 2-period model under uncertainty and prudence which expresses the general equilibrium equations in the approximation form written above (```uncertainty_SOE()```). The model under certainty equivalence serves the purpose of creating an initial *seed* to the numerical method performed below by the function ```root()```, whereas the second model is our model to solve using such numerical approximation. Notice that we have imposed the following values for the exogenous parameters of the model:

        \begin{align*}
        y_{1} &= 1.0\\
        y_{2}^{H} &= 1.05\\
        y_{2}^{L} &= 0.97\\
        p &= 0.55\\
        r &= 0.02
        \end{align*}

````python
import scipy.optimize as opt
from functools import partial
import numpy as np
import matplotlib.pyplot as plt

def certainty_eq_SOE(y,p=0.55,r=0.02):
    y1, y2H, y2L = y
    c1 = ((1+r)/(2+r))*(y1 + ((p*y2H + (1-p)*y2L)/(1+r)))
    c2H = (1+r)*(y1-c1) + y2H
    c2L = (1+r)*(y1-c1) + y2L
    ca1 = y1 - c1
    ca2H = y2H - c2H
    ca2L = y2L - c2L
    
    return c1, c2H, c2L, ca1, ca2H, ca2L

def uncertainty_SOE(c,y=[1.0,1.05,0.97],
                                 θ=1.0,
                                 r=0.02,
                                 p=0.55):
    c1, c2H, c2L = c
    y1, y2H, y2L = y
    ϵ_euler      = c1**(-θ) - (p*(c2H**(-θ)) + (1-p)*(c2L**(-θ)))
    ϵ_budgetH    = c2H - (1+r)*(y1-c1) - y2H
    ϵ_budgetL    = c2L - (1+r)*(y1-c1) - y2L
    ϵ = [ϵ_euler, ϵ_budgetH, ϵ_budgetL]
    
    return ϵ
````

2. ...

    - Plot the equilibrium values $c_{1},c_{2}^{H},c_{2}^{L},ca_{1}$ as a function of $\theta$...

        ... Now, solving our approximation model ```uncertainty_SOE()``` using ```opt.root()``` we solve for the equilibrium allocation for different values of $\theta$, i.e.

        $\theta\in\Theta = \{0.5,0.6,0.7,\ldots,2.0\}$

        and then we produce a graph for the different values of $\{c_{1},c_{2}^{H},c_{2}^{L},ca_{1}\}$ against the corresponding values of $\theta\in\Theta$. You  can see below that, as expected, the higher the relative risk aversion of the consumers, i.e. the higher the value of $\theta$, the less these individuals consume in the present to consume more in the future (regardless of the state of the world) by using their *precautionary savings*. This is finally reflected into the current account balance which shows that the higher the measure of relative risk-aversion, the less these consumers get into debt.

````python
p=0.55
y=[1.0,1.05,0.97]
Θ = np.linspace(0.5,2.0,16)
c_init = certainty_eq_SOE(y)[:3]
c1 = []
c2H = []
c2L = []
ca1 = []
ca2H = []
ca2L = []

for θ in Θ:
    ϵ_uncertainty = partial(uncertainty_SOE, θ=θ)
    c_solution = opt.root(ϵ_uncertainty, c_init)
    c1.append(c_solution.x[0])
    c2H.append(c_solution.x[1])
    c2L.append(c_solution.x[2])
    ca1.append(y[0] - c_solution.x[0])
    ca2H.append(y[1] - c_solution.x[1])
    ca2L.append(y[2] - c_solution.x[2])
    
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(Θ, c1, 'r')
axs[0, 0].set_xlabel(r'$\theta$')
axs[0, 0].set_ylabel(r'$c_{1}$')

axs[0, 1].plot(Θ, c2H, 'r')
axs[0, 1].set_xlabel(r'$\theta$')
axs[0, 1].set_ylabel(r'$c_{2}^{H}$')

axs[1, 0].plot(Θ, c2L, 'r')
axs[1, 0].set_xlabel(r'$\theta$')
axs[1, 0].set_ylabel(r'$c_{2}^{L}$')

axs[1, 1].plot(Θ, ca1, 'b')
axs[1, 1].set_xlabel(r'$\theta$')
axs[1, 1].set_ylabel(r'$ca_{1}$')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
````

3. **Example of an SOE with *No Precommitment***. Asuming that $b_{0}=0$, $\beta=\frac{1}{1+r}$ and $y_{1}<y_{2}$, consider the optimality conditions characterising the perfect foresight equilibrium of a 2-period SOE with *No Precommitment* in the international financial market :

    \begin{align*}
    u^{\prime}(c_1) &= u^{\prime}(c_{2}) + \chi\\
    d_{1}&\leq\frac{\phi y_{2}}{1+r}\\
    \chi\left(d_{1}-\frac{\phi y_{2}}{1+r}\right) &= 0\\
    c_{1} &= y_{1} + d_{1}\\
    c_{2} &= y_{2} - (1+r)d_{1}
    \end{align*}

* Assume that 

    \begin{align*}
    \phi=0.1\\
    r = 0.02
    \end{align*}

    and that preferences of the consumers of this economy are such that:
    
    $$ U_{1} =  \log{(c_{1})}+\beta\,\log{(c_{2})},$$

    and show how the perfect foresight equilibrium changes as the gap $\frac{y_{2}}{y_{1}}>1$ increases from 1.05 to 1.4.

    Although this can be easily done with pen and paper, below you can find a piece of code in python that first defines via the function ```noprecomitment()``` the perfect foresight equilibrium allocation for this economy summarised by the conditions above, and then shows in a graph how such allocation changes when we change the income gap.

    As can be seen in the graph below, when income in the first period is more than 25% lower than income in the second period, the SOE is no longer able to fund its flat consumption path, and the economy is forced to reduce consumption in the first period once the *no-default* condition and the ceiling on debt are binding. The evidence that the constraint is binding relies on the value of the lagrange multiplier $\chi$ which starts to take positive values.
     

````python
def noprecommitment(y,r=0.02,ϕ=0.1):
    y1, y2 = y
    assert y2 > y1, f"Error: The path of income must be increasing, i.e. y2 > y1"
    
    d = min(-(1/(2+r))*(y1-y2),ϕ*(y2/(1+r)))
    c1 = y1 + d
    c2 = y2 - (1+r)*d
    χ = (1/c1) - (1/c2)
    ca1 = -d
        
    return c1, c2, ca1, χ 
````

````python
y_gap = np.arange(1.05,1.45,0.01)
y1_values = 1/y_gap
c1_values = []
c2_values = []
ca1_values = []
χ_values = []

for y1 in y1_values:
   y = [y1, 1]
   c1, c2, ca, χ = noprecommitment(y)
   c1_values.append(c1)
   c2_values.append(c2)
   ca1_values.append(ca)
   χ_values.append(χ)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(y_gap, c1_values, 'r')
axs[0, 0].set_xlabel(r'$\frac{y_2}{y_1}$')
axs[0, 0].set_ylabel(r'$c_{1}$')
axs[0, 0].set_ylim(0.75, 1)

axs[0, 1].plot(y_gap, c2_values, 'r')
axs[0, 1].set_xlabel(r'$\frac{y_2}{y_1}$')
axs[0, 1].set_ylabel(r'$c_{2}$')
axs[0, 1].set_ylim(0.75, 1)

axs[1, 0].plot(y_gap, ca1_values, 'r')
axs[1, 0].set_xlabel(r'$\frac{y_2}{y_1}$')
axs[1, 0].set_ylabel(r'$ca_{1}$')

axs[1, 1].plot(y_gap, χ_values, 'b')
axs[1, 1].set_xlabel(r'$\frac{y_2}{y_1}$')
axs[1, 1].set_ylabel(r'$\chi$')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
````
